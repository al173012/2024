# **Dengan menggunakan jarak histogram (Eucledian) cari diantara citra2 dan citra 3 mana citra yang paling mirip dengan citra 1**


Berikut adalah hasil perbandingan jarak histogram (Euclidean) antara citra 1, citra 2, dan citra 3 berdasarkan histogram yang diberikan:

Jarak antara Citra 1 dan Citra 2:
Jarak Euclidean:  65796.3039

Jarak antara Citra 1 dan Citra 3:
Jarak Euclidean: 50244.2596

Berdasarkan perbandingan di atas, Citra 3 lebih mirip dengan Citra 1 daripada Citra 2 berdasarkan jarak histogram (Euclidean).

```python
import cv2
import math

# Load gambar (gantilah dengan path gambar yang sesuai)
gambar_a = cv2.imread('/content/citra1.png', cv2.IMREAD_GRAYSCALE)
gambar_b = cv2.imread('/content/citra2.png', cv2.IMREAD_GRAYSCALE)

# Hitung histogram dari gambar
hist_a = cv2.calcHist([gambar_a], [0], None, [256], [0, 256])
hist_b = cv2.calcHist([gambar_b], [0], None, [256], [0, 256])

# Hitung jarak Euclidean
sum_squared_diff = 0
for i in range(256):
    sum_squared_diff += (hist_a[i][0] - hist_b[i][0]) ** 2

euclidean_distance = math.sqrt(sum_squared_diff)
print(f'Jarak Euclidean antara citra 1 dan citra 2 adalah: {euclidean_distance:.4f}')

# Load gambar (gantilah dengan path gambar yang sesuai)
gambar_a = cv2.imread('/content/citra1.png', cv2.IMREAD_GRAYSCALE)
gambar_b = cv2.imread('/content/citra3.png', cv2.IMREAD_GRAYSCALE)

# Hitung histogram dari gambar
hist_a = cv2.calcHist([gambar_a], [0], None, [256], [0, 256])
hist_b = cv2.calcHist([gambar_b], [0], None, [256], [0, 256])

# Hitung jarak Euclidean
sum_squared_diff = 0
for i in range(256):
    sum_squared_diff += (hist_a[i][0] - hist_b[i][0]) ** 2

euclidean_distance = math.sqrt(sum_squared_diff)
print(f'Jarak Euclidean antara citra 1 dan citra 3 adalah: {euclidean_distance:.4f}')
```

# **Berapa besar moment warna order 1, energy, contras dan homogenitas dari citra**


```python
import numpy as np
from skimage.feature import greycomatrix, greycoprops

# Contoh matriks (gambar) dalam bentuk numpy array
matrix = np.array([[3, 2, 0, 3], [1, 2, 2, 0], [0, 2, 0, 0], [2, 3, 2, 3]],
                  dtype=np.uint8)

# Hitung matriks GLCM
distances = [1]  # Jarak antara piksel
angles = [0, np.pi/4, np.pi/2, 3*np.pi/4]  # Arah (horizontal, diagonal, vertikal)
glcm = greycomatrix(matrix, distances=distances, angles=angles, symmetric=True, normed=True)

# Hitung fitur
energy = greycoprops(glcm, 'energy')[0, 0]
contrast = greycoprops(glcm, 'contrast')[0, 0]
homogeneity = greycoprops(glcm, 'homogeneity')[0, 0]

print(f"Energy: {energy:.4f}")
print(f"Contrast: {contrast:.4f}")
print(f"Homogeneity: {homogeneity:.4f}")
```
